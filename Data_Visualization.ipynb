```python
#Question_No.5:Generate a synthetic dataset with correlated features. Visualize the correlation matrix 
#of a dataset using a heatmap.


#Answer:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with correlated features
# Let's create 3 features with some correlation
n_samples = 100
X1 = np.random.randn(n_samples)
X2 = X1 * 0.8 + np.random.randn(n_samples) * 0.2  # X2 is correlated with X1
X3 = X1 * 0.5 + np.random.randn(n_samples) * 0.5  # X3 is also correlated with X1, but less

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': X1,
    'Feature 2': X2,
    'Feature 3': X3
})

# Compute the correlation matrix
corr_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)

# Adding labels and title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
#Q2.A Using the sales data, generate a heatmap to visualize the variation in sales across
#different months and days.


#Answer: 2. Heatmap for Sales Data:
#The heatmap will visualize the variation in sales across different months and days. We will aggregate 
#the sales data by Month and Day, and then plot the heatmap.

# Sales Data
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df_sales = pd.DataFrame(data_sales)

# Pivot the data to create a matrix for heatmap visualization
sales_pivot = df_sales.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum')

# Create a heatmap to show variation in sales across months and days
plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, annot=True, cmap='YlGnBu', linewidths=0.5, fmt='g', cbar_kws={'label': 'Sales'})

# Adding labels and title
plt.title('Variation in Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Show the plot
plt.show()
```

```python
#Question_No.3: Using the sales data, generate a heatmap to visualize the variation in sales across
#different months and days.
#np.random.seed(20) 
#data = { 
#    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100), 
#    'Day': np.random.choice(range(1, 31), 100), 
 #   'Sales': np.random.randint(1000, 5000, 100) 
#} 
#df = pd.DataFrame(data)

#Answer:
import numpy as np
import pandas as pd
import plotly.express as px

# Set random seed for reproducibility
np.random.seed(20)

# Generate sales data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Pivot the data to create a matrix of sales values
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create the heatmap
fig = px.imshow(
    heatmap_data,
    labels={'x': 'Month', 'y': 'Day', 'color': 'Average Sales'},
    x=heatmap_data.columns,
    y=heatmap_data.index,
    color_continuous_scale='Viridis'
)

# Update layout for better visualization
fig.update_layout(
    title='Sales Heatmap by Month and Day',
    xaxis_title='Month',
    yaxis_title='Day',
    xaxis=dict(side='top')
)

# Show the plot
fig.show()
```

```python
#Question_No.1:Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the
#sine of x.


#Answer:
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate data for sine wave
x = np.linspace(0, 10, 100)  # x-values from 0 to 10
y = np.sin(x)  # y-values as sine of x

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y', 
           width=800, height=400)  # Use 'width' and 'height' instead of 'plot_width' and 'plot_height'

# Plot the sine wave as a line
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the plot in the notebook
output_notebook()  # This ensures the plot will appear in the notebook
show(p)
```

```python
#Question_No.2:Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes
#and colors for the markers based on the 'sizes' and 'colors' columns.


#Answer:
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(42)
n = 100  # Number of points
x = np.random.rand(n) * 10  # Random x values between 0 and 10
y = np.random.rand(n) * 10  # Random y values between 0 and 10
sizes = np.random.randint(10, 100, n)  # Random sizes between 10 and 100
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)  # Random colors

# Create a DataFrame for the data
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(df)

# Create a scatter plot
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Scatter plot with varying sizes and colors
p.scatter(x='x', y='y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

# Show the plot in the notebook
output_notebook()  # This ensures the plot will appear in the notebook
show(p)
```

```python
#Question_No.3:Generate a Bokeh bar chart representing the counts of different fruits using the 
#following dataset.
#fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] 
#counts = [20, 25, 30, 35]

#Answer:
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource from the data
data = {
    'fruits': fruits,
    'counts': counts
}

source = ColumnDataSource(data)

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', 
           width=800, height=400)

# Create a bar chart
p.vbar(x='fruits', top='counts', width=0.6, source=source, legend_field='fruits', color="green", alpha=0.7)

# Show the plot in the notebook
output_notebook()  # This ensures the plot will appear in the notebook
show(p)
```

```python
#Question_No.4: Create a Bokeh histogram to visualize the distribution of the given data.
#data_hist = np.random.randn(1000) 
#hist, edges = np.histogram(data_hist, bins=30)

#Answer:
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Generate random data
data_hist = np.random.randn(1000)  # Data following a normal distribution

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create the left and right edges
left_edges = edges[:-1]
right_edges = edges[1:]

# Create a ColumnDataSource from the histogram data
source = ColumnDataSource(data={'hist': hist, 'left': left_edges, 'right': right_edges})

# Create a Bokeh figure
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', 
           width=800, height=400)

# Plot the histogram as vertical bars
p.quad(top='hist', bottom=0, left='left', right='right', 
       source=source, fill_color="green", line_color="black", alpha=0.7)

# Show the plot in the notebook
output_notebook()  # This ensures the plot will appear in the notebook
show(p)
```

```python
#Question_No.5:Create a Bokeh heatmap using the provided dataset.
#data_heatmap = np.random.rand(10, 10) 
#x = np.linspace(0, 1, 10) 
#y = np.linspace(0, 1, 10) 
#xx, yy = np.meshgrid(x, y)

#Answer:
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, LinearColorMapper
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)  # 10x10 random values between 0 and 1
x = np.linspace(0, 1, 10)  # x-axis values from 0 to 1 (10 points)
y = np.linspace(0, 1, 10)  # y-axis values from 0 to 1 (10 points)
xx, yy = np.meshgrid(x, y)  # Create a mesh grid for x and y

# Flatten the mesh grid and the heatmap data
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': x_flat, 'y': y_flat, 'z': z_flat})

# Create a LinearColorMapper
color_mapper = LinearColorMapper(palette=Viridis256, low=min(z_flat), high=max(z_flat))

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=800, height=800)

# Create the heatmap using rectangles (p.rect)
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, 
       color={'field': 'z', 'transform': color_mapper}, 
       line_color='white', alpha=0.7)

# Show the plot in the notebook
output_notebook()  # This ensures the plot will appear in the notebook
show(p)
```
